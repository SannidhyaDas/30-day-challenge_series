```python
#Find the sum of a 3 digit number given by the user
num = int(input("Enter a 3-digit number: "))
if 100 <= num <= 999:
    digit1 = num // 100
    digit2 = (num // 10) % 10
    digit3 = num % 10
    total = digit1 + digit2 + digit3
    print("The sum of the digits is:", total)
else:
    print("Please enter a valid 3-digit number.")    
```

```python
#printing a multiplication table of a given number with proper formatting
start_time = time.time()
num = int(input("Enter a number to print its multiplication table: "))
print(f"Multiplication Table of {num}:")
for i in range(1, 11):
    print(f"{num:2} x {i:2} = {num * i:3}")
    

end_time = time.time()
execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")    
```

```python
#Guess the number game
import random
number_to_guess = random.randint(1, 100)
attempts = 0
while True:
    user_guess = int(input("Guess a number between 1 and 100: "))
    attempts += 1
    if user_guess < number_to_guess:
        print("Too low! Try again.")
    elif user_guess > number_to_guess:
        print("Too high! Try again.")
    else:
        print(f"Congratulations! You've guessed the number {number_to_guess} in {attempts} attempts.")
        break
   
```

```python
#Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% 
# per year. You have to write a program to find out the population at the end of each of the upcoming 10 years.

start_time = time.time()

population = 10000
growth_rate = 0.10
for year in range(1, 11):
    population += population * growth_rate
    print(f"Year {year}: Population = {int(population)}")

end_time = time.time()
execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")
```

```python
#Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% 
# per year. You have to write a program to find out the population at the end of each of the last 10 years.

start_time = time.time()

population = 10000
growth_rate = 0.10
for year in range(10, 0, -1):
    print(f"Year {year}: Population = {int(population)}")
    population -= population * growth_rate

end_time = time.time()
execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")
```

```python
#printing Fibonacci series up to n terms
n = int(input("Enter the number of terms in Fibonacci series: "))
a, b = 0, 1
count = 0
while count < n:
    print(a, end=' ')
    a, b = b, a + b
    count += 1
```

```python
import math
num = int(input("You want your series sum upto: "))
sum_series = 0
for i in range(num):
    sum_series += i/math.factorial(i)
print(f"The sum of the series up to {num} is: {sum_series}")
```

```python
num = int(input("You want your series sum upto: "))
sum_series = 0
count=1
fact=1
while count < num+1:
    fact=fact*count
    sum_series += count/fact
    count += 1
print(f"The sum of the series up to {num} is: {sum_series}")    
```
